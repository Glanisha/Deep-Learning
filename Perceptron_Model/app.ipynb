Implement and get an AND gate with the percentron model assuming the weights. If the output is not the desired output get an error and modify the weights so that we implement a learning mechanism. 



```python
# Algo by Miss
# Initialize weights and bias to zero

# Repeat for a fixed number of epochs:
#     a. For each training sample x, y:
#         i. Compute weighted sum: z = w·x + b
#         ii. Apply activation: y_pred = step(z)
#         iii. Compute error: error = y - y_pred
#         iv. Update weights: w = w + α × error × x_i
#         v. Update bias: b = b + α × error
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
class Perceptron: 
    def __init__(self, inputSize, learningRate):
        self.weights=np.random.uniform(-1, 1, inputSize)
        self.alpha=learningRate
        self.bias=0.0

    def f_function(self, z):
        if (z>=0): return 1
        else: return 0

    def computation_of_weighted_sum(self, x):
        z=0.0
        for i in range(len(x)):
            z+=x[i]*self.weights[i]
        z+=self.bias
        return z
    
    def predict(self, x):
        z=self.computation_of_weighted_sum(x)
        return self.f_function(z)
    
    def update_weights_and_bias_for_error(self, x, error):
        for i in range(len(self.weights)):
            self.weights[i]+=self.alpha*error*x[i]
        self.bias+=self.alpha*error
        
    def training(self , X, Y, epoch):
        for e in range(epoch):
            print(f"\nEpoch number : {e+1}")
            for i in range(len(X)):
                x=X[i]
                y=Y[i]
                z=self.computation_of_weighted_sum(x)
                y_pred=self.f_function(z)
                error=y-y_pred
                self.update_weights_and_bias_for_error(x, error)
                print(f"x: {x}, y: {y}, z: {z:.2f}, y_pred: {y_pred}, error: {error}, weights: {self.weights}, bias: {self.bias:.2f}")
```

```python
def main():
    X = [
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ]
    Y = [0, 0, 0, 1]
    model = Perceptron(inputSize=2, learningRate=0.1)
    model.training(X, Y, epoch=20)
if __name__ == "__main__":
    main()
```
